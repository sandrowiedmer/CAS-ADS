## 0. About data management

It can be a camera, a temperature sensor, a telescope, a microscope (with a camera), a microphone, a particle detector etc. Normally the data is then digitised, maybe preprocessed and written to some media in a certain format, e.g. as a comma separated value (csv) file to a hard disk. This part of the data management is normally taken care of by engineers.

Data may also be collected from all sorts of databases, so data already collected somehow. Time series of financial data, customers, passengers, facebook likes, twitter tweets etc. This is data which is normally already on a media with some interface for access, e.g. paper to be read by a camera, a file on youtube, a table on wikipedia etc. We will look at some ways to collect such data. 1 = least exposed (e.g. animal sleeps in a well-protected den)\
5 = most exposed
* Danger: overall danger index (1-5) based on the above two indices and other information
1 = least danger (from other animals)\
5 = most danger (from other animals)

### Exercise 4
Investigate what ```how='inner'``` does and what other options exist.



The expression ```how='inner'``` performs an inner join between the two dataframes meaning that only species existing in both dataframes end up in the merged dataframe.

Other options are:
- ```how='left'```: includes all rows from the left dataframe, and matching rows from the right dataframe.

- ```how='right'```: includes all rows from the right dataframe, and matching rows from the left dataframe.

- ```how='outer'```: includes all rows from both data frames, with NaNs where there are no matches.

````python
import re
from collections import Counter
from sentence_transformers import SentenceTransformer
from sklearn.cluster import AgglomerativeClustering

# identifiers
subjects = ["Religion", "Wirtschaft & Recht", "Physical Education", "Philosophie", "BG",
 "Wirtschaft und Recht", "Mathe, Statistik, alle fächer die ich bei frau mühlemann gehabt hätte",
 "Musik", "Sport", "Englisch", "Mathematik", "Politikwissenschaft", "Geschichte", "Chemie",
 "*Passerelle: Physik", "WR", "WIRE", "Wirtschaft", "Physik", "Biologie", "Psychologie und Pädagogik",
 "Bg", "Mathematik \r\nWirtschaft", "Deutsch", "Wirtschaft + Recht", "Bildnerisches Gestalten", "bwl",
 "Spanisch", "Wr", "English", "Psychologie", "Math", "sport", "math", "Geografie", "Sport und Musik",
 "Italienisch", "Geschichte, Wirtschaft und Recht", "Sport und WR", "Franzoesisch", "spanisch",
 "Rätoromanisch", "SWIR", "(Bio-)Chemie", "VWL", "SF Bio Chemie", "Geo", "GESCHICHTE", "deutsch",
 "Theater", "Art", "Bildnerisches gestalten", "Politische Bildung", "Kunst", "Mathe", "mathematik",
 "hatte keins", "Englisch, Mathematik", "SF Musik", "Geographie", "W&R und Sport", "EWR",
 "Wirtschaft Geschichte", "Digitale Kommunikation", "Technisches Gestalten", "Musik Kunstfach",
 "Französisch", "Biologie und Chemie", "Religion, Psychologie", "englisch", "Geschichte und Wirtschaft",
 "Französich", "geschichte", "sport, geschichte, Wirtschaft,", "gar keins", "astrophysik",
 "Wiertschaft und Recht", "Informatik", "Sport, Geschichte", "Geschichte und Recht", "keines", "BWL",
 "Schwerpunktfach Musik", "Latein", "Mein Lieblingsfach war Fremdsprache.", "Anwendungen der Mathematik"]

# ----------------
# Helpers
# ----------------
def normalize_text(text: str) -> str:
    """Clean and lowercase a subject name."""
    t = text.lower()
    t = re.sub(r"[^a-zäöüß ]", " ", t)
    t = re.sub(r"\s+", " ", t).strip()
    return t

def propose_cluster_name(items: list[str]) -> str:
    """Pick best representative name for a cluster."""
    normalized = [normalize_text(x) for x in items]
    counts = Counter(normalized)
    # pick most common normalized term
    most_common, freq = counts.most_common(1)[0]
    if freq > 1:
        return most_common
    # else: fallback to shortest original item
    return min(items, key=len)

# ----------------
# 1. Encode
# ----------------

model = SentenceTransformer("paraphrase-multilingual-MiniLM-L12-v2")
#model = SentenceTransformer("sentence-t5-xl")

embeddings = model.encode(subjects)

# ----------------
# 2. Cluster
# ----------------
clustering = AgglomerativeClustering(n_clusters=None, distance_threshold=1.5)
labels = clustering.fit_predict(embeddings)

# ----------------
# 3. Build mapping & propose names
# ----------------
cluster_map = {}
for subj, label in zip(subjects, labels):
    cluster_map.setdefault(label, []).append(subj)

for cluster_id, items in cluster_map.items():
    cluster_name = propose_cluster_name(items)
    print(f"\nCluster {cluster_id} → Proposed name: {cluster_name} ({len(items)} items)")
    for i in items:
        print(f"  - {i}")
````
